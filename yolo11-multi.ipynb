# yolo11-multi (postdata)

```python
import os
os.environ["MKL_SERVICE_FORCE_INTEL"] = "1"

from ultralytics import YOLO

# Load a model
# model = YOLO("my_yolo11.yaml")  # build a new model from YAML
model = YOLO("yolo11-gconv.yaml")  # load a pretrained model (recommended for training)
# model = YOLO("my_yolo11.yaml").load("weights/yolo11n.pt")  # build from YAML and transfer weights

# Train the model
results = model.train(data="mydata-linux-processed.yaml", epochs=100, imgsz=640, batch=96, device = [0,1,2,3], workers=32, cache='ram')
```
